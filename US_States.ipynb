```python
                                # US DATA OF STATES
import numpy as np                                 
import pandas as pd
pop = pd.read_csv('C:\\Users\\Admin\\Documents\\Python practice\\data-USstates-master\\state-population.csv')
areas = pd.read_csv('C:\\Users\\Admin\\Documents\\Python practice\\data-USstates-master\\state-areas.csv')
abbrevs = pd.read_csv('C:\\Users\\Admin\\Documents\\Python practice\\data-USstates-master\\state-abbrevs.csv')
print(pop.head())
print(areas.head())
print(abbrevs.head())            
print('----------------------------------Step1: Many to one Merge---------------------------------------------------')
print('This step will give us the full state name from abbrev dataframe and will be added to pop DataFrame')
print('Merge will be on basis of state column of pop DataFrame and abbreviation column of abbrev')
merged = pd.merge(pop, abbrevs, how='outer', left_on='state/region', right_on='abbreviation').drop('abbreviation', axis=1)
print(merged.head())
print('Check for any mis-match: It can be done by checking if there is any NaN value present in any row')
print(merged.isnull().any())
print('It looks like there are some null values in population; we shall find it out by boolean masking!!')
print(merged[merged['population'].isnull()].head())
print('As the states column is also showing True in previous table: which means that some States name are missing')
merged.loc[merged['state'].isnull(), 'state/region'].unique()
print('PR and USA abbrevations full names are missing, So we will add them..')
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Peurto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'Unites States'
print('After adding names of states we shall check it')
print(merged.isnull().any())
print('Now there are no NaN states, but there are NaN population beacuse that data is not provided in the Csv file.')
print('------------------------------------Step2: Merge Area DataFrame to merge DataFrame------------------------------------')
final = pd.merge(merged, areas, on='state', how='left')
print(final.head())
print('Now we should check if there are some mismathch in final dataframe.')
print(final.isnull().any())
print('As seen there is some mismatch in area, so we will find it out....')
print(final['state'][final['area (sq. mi)'].isnull()].unique())
print('We will remove these two states')
final.dropna(inplace=True)
print(final.isnull().any())
print('There are no NaN values in the final DataFrame')
data2010 = final.query("year == 2010 & ages == 'total'")
print(data2010.head())
print('All the values of year 2010 and ages total')
data2010.set_index('state', inplace=True)
print(data2010.head())
density = data2010['population']/data2010['area (sq. mi)']
print('=================================================POPULATION DENSITY=================================================')
print(density)
print('Thus the population density of 2010 year is calculated.............')
```
